```python
import joblib
import m2cgen as m2c

# 1. Load model
model = joblib.load('../models/runform_model.pkl')

# 2. Konversi ke JAVA (ini yang didukung m2cgen)
model_java_code = m2c.export_to_java(model)

# 3. Simpan sebagai file .java
with open('../models/RunformPredictor.java', 'w') as f:
    f.write(model_java_code)

print("Model berhasil dikonversi ke JAVA!")
```
